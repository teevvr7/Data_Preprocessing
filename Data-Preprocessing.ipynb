### Stage 2
Data Transformation
- If data is continuous = Standard Scaler, MinMaxScaler, Robust Scaler
- If data is discrete = Label Encoder, One Hot Encoder


```python
weather_data.head()
```

```python
weather_data_copy = weather_data.copy()
```

#### 1.Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le_encoder=LabelEncoder()
```

```python
weather_data_copy['Weather'] = le_encoder.fit_transform(weather_data_copy['Weather']) 
```

#### 2.One Hot Encoder

1. Pandas - pd.get_dummies()
2. sklearn - OneHotEncoder()

```python
weather_data_copy2 = weather_data.copy()
```

```python
weather_data_copy2.head()
```

```python
weather_data_copy2=pd.get_dummies(data= weather_data_copy2,columns=['Weather'],dtype = 'int',drop_first = True)
weather_data_copy2
```

```python
weather_data_copy2.dtypes
```

```python
weather_data_copy3 = weather_data.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)

ohe_fit = ohe.fit_transform(weather_data_copy3[['Weather']])

encoded_copy3 = pd.DataFrame(
    ohe_fit,
    columns=ohe.get_feature_names_out(['Weather'])
)

print(encoded_copy3)
```

```python
encoded_copy3
```

```python
del weather_data_copy3['Weather']
```

```python
weather_data_copy3=pd.concat([weather_data_copy3,encoded_copy3],axis=1)
```

```python
weather_data_copy3.head()
```

## How to choose between Label Encode and One Hot Encoder

Input Feature:

Based on whether it is Parametric or a Non Parametric Model, we will choosing between LE or OHE

- For Parametric Models : Better to go with OHE
- For Non-Parametric Models : always go with Label Encoding

Output feature :

- Always the output feature, it must be Label Encoder.
